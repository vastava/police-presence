```python
import pandas as pd
import numpy as np

offenses = pd.read_csv('offenses_city.csv')
race = pd.read_csv('city_census_race.csv')
police = pd.read_csv('police_city.csv')
income = pd.read_csv('city_median_income.csv')
```

```python
cities = np.array(race['City'])
states = np.array(race['State'])
cities
cities
```

```python
from string import digits

remove_digits = str.maketrans('', '', digits)

offenses['City'] = offenses['City'].apply(lambda x: str(x).translate(remove_digits))
police['City'] = police['City'].apply(lambda x: x.translate(remove_digits))
offenses['Abbrev'] = offenses['Abbrev'].apply(lambda x: str(x))
```

```python
police_vals = []
offenses_vals = []
incomes_vals = []
for i in range(len(cities)):
    city = cities[i]
    state = states[i]
    print(city)
    print(state)
    police_row = police[(police.Abbrev == state) & (police.City == city)]
    offenses_row = offenses[(offenses.Abbrev == state) & (offenses.City == city)]
    income_row = income[(income.State == state) & (income.City == city)]
    if city == "Albuquerque" or city == "Jackson":
        temp = 0
    else:
        temp = police_row.Per_1k.values[0]
    police_vals.append(temp)
    if city == "Raleigh" or city == "Louisville" or city == "Winston-Salem" or state == "IA":
        temp2 = 0
    else:
        temp2 = offenses_row.Per_1k.values[0]
    offenses_vals.append(temp2)
    temp3 = income_row.Median_Income.values[0]
    incomes_vals.append(temp3)
```

```python
len(police_vals)
```

```python
race["police_per_1k"] = police_vals
race["offenses_per_1k"] = offenses_vals
race["median_income"] = incomes_vals
race
```

```python
race.to_csv('race_comparison.csv')
```
